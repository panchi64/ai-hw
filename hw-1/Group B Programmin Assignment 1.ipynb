```python
import timeit
import numpy as np
import matplotlib.pyplot as plt

# Function to perform array multiplication using pure Python
def multiply_arrays_python(arr1, arr2):
    result = []
    for i in range(len(arr1)):
        sum_val = 0
        for j in range(len(arr2[0])):
            sum_val += arr1[i] * arr2[i][j]
        result.append(sum_val)
    return result

# Function to conduct the experiment and measure execution time
def run_experiment(dimensions):
    python_times = []
    numpy_times = []

    for dim in dimensions:
        # Generate random arrays
        arr1 = np.random.rand(dim)
        arr2 = np.random.rand(dim, dim)

        # Measure execution time for pure Python
        python_time = timeit.timeit(lambda: multiply_arrays_python(arr1.tolist(), arr2.tolist()), number=1)
        python_times.append(python_time)

        # Measure execution time for NumPy
        numpy_time = timeit.timeit(lambda: np.dot(arr1, arr2), number=1)
        numpy_times.append(numpy_time)

    return python_times, numpy_times

# Defined dimensions for the experiment
small_dimensions = [1,10]
medium_dimensions = [10,100]
large_dimensions = [100,1000]

# Run the experiment for all array sizes
small_python_times, small_numpy_times = run_experiment(small_dimensions)
medium_python_times, medium_numpy_times = run_experiment(medium_dimensions)
large_python_times, large_numpy_times = run_experiment(large_dimensions)

# Visualize the results for small arrays
plt.figure(figsize=(10, 6))
plt.plot(small_dimensions, small_python_times, label='Pure Python (Small)')
plt.plot(small_dimensions, small_numpy_times, label='NumPy (Small)')
plt.xlabel('Dimension')
plt.ylabel('Time (seconds)')
plt.title('Array Multiplication: Pure Python vs NumPy (Small Arrays)')
plt.legend()
plt.grid(True)
plt.show()

# Visualize the results for medium arrays
plt.figure(figsize=(10, 6))
plt.plot(medium_dimensions, medium_python_times, label='Pure Python (Medium)')
plt.plot(medium_dimensions, medium_numpy_times, label='NumPy (Medium)')
plt.xlabel('Dimension')
plt.ylabel('Time (seconds)')
plt.title('Array Multiplication: Pure Python vs NumPy (Medium Arrays)')
plt.legend()
plt.grid(True)
plt.show()

# Visualize the results for large arrays
plt.figure(figsize=(10, 6))
plt.plot(large_dimensions, large_python_times, label='Pure Python (Large)')
plt.plot(large_dimensions, large_numpy_times, label='NumPy (Large)')
plt.xlabel('Dimension')
plt.ylabel('Time (seconds)')
plt.title('Array Multiplication: Pure Python vs NumPy (Large Arrays)')
plt.legend()
plt.grid(True)
plt.show()
```
